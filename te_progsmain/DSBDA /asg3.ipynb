```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
iris = load_iris()
iris.keys()
```

```python
df = pd.DataFrame(iris.data , columns=iris.feature_names)
df["Label"] = iris.target
```

```python
df.head()
```

```python
df.mean()
```

```python
df.mode()
```

```python
df.median()
```

```python
setosa = df[["sepal length (cm)", "sepal width (cm)", "petal width (cm)"]].groupby(df["petal length (cm)"])
```

```python
setosa.head()
```

```python
mean = setosa.mean()
print(mean)
```
